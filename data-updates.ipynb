## For topic data (.json)
Only require the top and second level of topics. 
The **top level topics** directly use topic `name`, the **second level topics** use their codes as `short` and full name as `name`.

```python
import json
```

```python
# json file structure summary
def summarise(obj, indent=0):
    pad = '  ' * indent
    if isinstance(obj, dict):
        for k, v in obj.items():
            print(f"{pad}{k}: {type(v).__name__}")
            if isinstance(v, (dict, list)):
                summarise(v, indent + 1)
    elif isinstance(obj, list):
        print(f"{pad}[list of {len(obj)} items]")
        if obj:
            summarise(obj[0], indent + 1)

with open('pax_topics.json') as f:
    topics = json.load(f)

with open('hierarchy.json') as f:
    hierarchy = json.load(f)

# Full topic structure
# summarise(topics)

# Topics structure used in the visualization
summarise(hierarchy)
```

## For PA-X data (.csv)

```python
import pandas as pd

df_old = pd.read_csv('pax_data_v9.csv')
df_new = pd.read_csv('pax_data_v10.csv')
```

```python
cols_old = set(df_old.columns)
cols_new = set(df_new.columns)

only_in_old_file = cols_old - cols_new
only_in_new_file = cols_new - cols_old
common = cols_old & cols_new

print(f"Columns only in old file: {only_in_old_file}")
print(f"Columns only in new file: {only_in_new_file}")
print(f"Common columns: {common}")
```
